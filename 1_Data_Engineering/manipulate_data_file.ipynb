```python
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
data_path = '../data/MunE19_c.csv'
```

```python
# Leer el archivo CSV sin índice
df = pd.read_csv(data_path, sep=';', encoding='ISO-8859-1')

# Crear una nueva columna de índice que comience en 0
df.insert(0, 'index', range(0, len(df)))

# Obtener el nombre del archivo original sin la extensión
file_name, file_extension = os.path.splitext(data_path)

# Crear el nombre del archivo de salida añadiendo "_mod"
data_mod_path = f"{file_name}_mod{file_extension}"

# Guardar el archivo CSV con el nuevo índice
df.to_csv(data_mod_path, index=False)

print(f"✨ The modified CSV file has been saved to: {data_mod_path}\n")

print(df.head())

print(df.shape)

print(df.info())

print(df.describe(include='all'))
```

```python
# Define the path to the original CSV file (assuming data_mod_path is already defined)
csv_file_path = data_mod_path  

# Extract the file name and extension from the original CSV file path
file_name, file_extension = os.path.splitext(os.path.basename(csv_file_path))

# Define the new CSV file path with 'municipios' added to the file name
new_file_name = f"{file_name}_municipios.csv"  # Exclude original extension
new_csv_file_path = os.path.join(os.path.dirname(csv_file_path), new_file_name)

# Select the first 5 columns
df_first_5_columns = df.iloc[:, :5]  # Selecting all rows and first 5 columns

# Save the selected columns to a new CSV file
df_first_5_columns.to_csv(new_csv_file_path, index=False)

print(f"✨ New CSV file '{new_csv_file_path}' created with the first 5 columns.\n")

print(df_first_5_columns.head())

print(df_first_5_columns.shape)

print(df_first_5_columns.info())

print(df_first_5_columns.describe(include='all'))
```
